## Clustering

Since we do not have any labels, I consider the following options:
-  Use <b>heuristics</b>: (if "analytics in text" => topic='analytics')
-  Try to <b>cluster</b> the texts and find the cluster that corresponds to the analytics topic.
-  Label some examples and use <b>active learning</b> to leave the model learn the rest for you.

### Heuristics

```python
# Define analytics keywords
analytics_keywords = ['analytics', 'data science', 'artificial intelligence', 'ai', 'ki', 'künstliche intelligenz']

# Add word counts by using a count veectorizer from sklearn
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(vocabulary=analytics_keywords)

# Get text as a list for vectorizer
texts = clean_df.cleaned_text.to_list()
count_matrix = vectorizer.fit_transform(texts)
```

```python
# Convert to pandas to handle results
count_df = pd.DataFrame(count_matrix.todense(), columns=analytics_keywords)

# Create a column that sums up occurrencies
count_df['total'] = count_df[analytics_keywords].sum(axis=1)

# Set up a threshold and a label
threshold = 1
count_df['label_heuristics'] = count_df['total'].apply(lambda x: 1 if x>threshold else 0)
```

```python
# Append this dataset to the main one
label_df = pd.concat([clean_df, count_df[['label_heuristics']]], axis=1)
```

### Clustering

#### Create embeddings from text

```python
# Use tf-idf embeddings since we are using BOW
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(
    ngram_range=(1, 1),
#     min_df=0.1,
    max_df=0.8,
    max_features=5000
)

# Get text as a list for vectorizer
texts = df.cleaned_text.to_list()
tf_idf_matrix = vectorizer.fit_transform(texts)
```

#### Cluster results

##### Elbow method + k-means

We can also try to find the optimal number of cluster by using the elbow method.

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

# Define range of n of clusters
cluster_range = np.arange(start=2, stop=30, step=1)

# Save parameters here
distortions = []
inertias = []

# Iterate over the range and calculate distorsion and inertia
for i in cluster_range:
    # Fit
    k_means_i = KMeans(n_clusters=i)
    k_means_i.fit(tf_idf_matrix)
    # Append inertia (squared distance of each sample to its center)
    inertias.append(k_means_i.inertia_)
    
```

```python
# Plot elbow plot
plt.plot(
    cluster_range,
    inertias,
    color='darkblue',
    marker='o',
)
plt.xlabel('# clusters')
plt.ylabel('Inertia')
plt.title('The elbow method recommends 3 clusters.');
```

Now, use this number of cluster for the k-means algorithm.

```python
# Fit and transform tf_idf
elbow_cluster = 8
k_means = KMeans(n_clusters=elbow_cluster).fit(tf_idf_matrix)

# Append labels
df['k_means_cluster'] = k_means.labels_
```

#### Use UMAP to visualize the results in 2D

```python
import umap

# Define mapper
reducer = umap.UMAP(
    n_neighbors=50,
    n_components=2
)

# Transform
umap_embeddings = reducer.fit_transform(tf_idf_matrix)
```

```python
from sklearn.cluster import DBSCAN

# Fit cluster to data
clustering = DBSCAN(eps=0.5, min_samples=3).fit(umap_embeddings)

# Append labels to dataset
df['dbscan_cluster'] = clustering.labels_
```

##### Visualize DBSCAN

```python
# PLot
import seaborn as sns
plt.scatter(
    umap_embeddings[:, 0],
    umap_embeddings[:, 1],
    c=[sns.color_palette()[x] for x in df['dbscan_cluster'].tolist()]
)
plt.title('UMAP projection with DBSCAN clusters');
```

##### Visualize K-Means

```python
# PLot
import seaborn as sns
plt.scatter(
    umap_embeddings[:, 0],
    umap_embeddings[:, 1],
    c=[sns.color_palette(n_colors=30)[x] for x in df['k_means_cluster'].tolist()]
)
plt.title('UMAP projection with K-Means clusters');
```

## Model with heuristics

Since we do not have any labels and the clustering could not separate analytics topics from the other topics, I will use heuristics to determine whether a text is analytics related.

### Create and save model

```python
from sklearn.pipeline import Pipeline
from predictor.Predictor import Predictor
from preprocessor.Preprocessor import Preprocessor

# Define keywords to search for
analytics_keywords = [
    'analytics', 'data science', 'data analytics', 'artificial intelligence', 'ai', 'ki', 'künstliche intelligenz'
]

# Load model
# nlp = spacy.load('en_core_web_sm')

# Add munich re to stopwords
# nlp.Defaults.stop_words |= {"munich", "re", "munichre"}

# Create pipeline
pipe = Pipeline(
    [
        ('preprocess', Preprocessor(nlp=None)),
        ('count_vectorizer', CountVectorizer(vocabulary=analytics_keywords, ngram_range=(1,2))),
        ('predictor', Predictor(vocabulary=analytics_keywords, threshold=1))
    ]
)
```

```python
pipe.transform(
    [
        'I love data analytics!',
        'ai is the future of data science'
    ]
)
```

```python
from sklearn.base import BaseEstimator
import py3langid as langid

class Model(BaseEstimator):
    def __init__(self, pipeline, lang=None):
        self.lang = lang
        self.pipeline = pipeline

    def fit(self, X, y=None):
        return self

    def predict(self, X):
        # So that it also works with lists, we need this step.
        if isinstance(X, list):
            X = pd.DataFrame(X, columns=['text'])
        # Get language
        lang_col = X['text'].apply(lambda x: langid.classify(x)[0]).tolist()
        
        # Transform
        X = self.pipeline.transform(X)
        
        # Overwrite
        results = pd.DataFrame(
            list(zip(lang_col, X)), columns=['lang', 'prediction']
        )
        results['prediction'] = results.apply(
            lambda x: x['prediction'] if x['lang']==self.lang else -1, axis=1
        )
        return results['prediction']
```

```python
model = Model(
    pipeline=pipe,
    lang='en'
)
```
